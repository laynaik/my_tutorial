```python
n_points=100
```

```python
n_points
```

```python
import numpy as np
```

```python
x=np.linspace(0,1,n_points)
```

```python
x
```

```python
z=np.sin(2*np.pi*x)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,0))
```

```python
plt.plot(x,z,'g-')
```

```python
eta=.4*np.random.normal(0,1,n_points)
```

```python
eta
```

```python
plt.plot(x,eta,'r')
```

```python
t=z+eta
plt.plot(x,z,'g',x,t,'bo')
plt.legend(('sine wave','noisy data'))
```

```python
sample_points=10
import numpy as np
x_sample=np.linspace(0,1,sample_points)
eta_sample=.2*np.random.normal(0,1,sample_points)
t_sample=np.sin(2*np.pi*x_sample)+eta_sample
import matplotlib.pyplot as plt
plt.plot(x_sample,t_sample,'bo',x,z,'g-')
plt.legend(('noisy sample point','sine wave'))
         
         

```

# polynomial curve fitting 

```python
m=3
coeff=np.polyfit(x_sample,t_sample,m)
print(coeff)
```

```python
p=np.poly1d(coeff)
print(p)
```

```python
y_pred=p(x_sample)
```

```python
plt.plot(x_sample,t_sample,'bo',x_sample,y_pred,'r-')
plt.legend(('training data','model with m=3'))
```

```python
err=0.5*(np.sum(y_pred-t_sample)**2)
print(err)
```

```python
n=5
x_new=np.linspace(0,1,n)
eta_new=.2*np.random.normal(0,1,n)
y_new=np.sin(2*np.pi*x_new)+eta_new

y_new=p(x_new)

plt.plot(x_sample,t_sample,'bo',x,z,'g',x_new,y_new,'ro')
plt.legend(('taining data','model','testing data'))
```

```python
z3=np.polyfit(x_sample,t_sample,3)
p3=np.poly1d(z3)
y3=p3(x_sample)

z9=np.polyfit(x_sample,t_sample,9)
p9=np.poly1d(z9)
y9=p9(x_sample)

plt.plot(x_sample,t_sample,'bo',x_sample,y3,'g',x_sample,y9,'r')
plt.legend(('sample data','m=3','m=9'))
```
